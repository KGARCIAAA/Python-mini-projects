Use the [pokeAPI](https://pokeapi.co/)

1. Find out information about the Pokemon "tyranitar":
- What is its type(s)?
- What is its abilities?

2. How many pokemon are "dragon" type?

You must use python and apply the GET method via their API to display this information.

```python
import requests #need to request data from the API
import ratelim
import json
import pandas as pd
import pprint
```

```python
# here I am creating the response variable and requesting
# data from the pokeAPI server using the requests.get() function
url = 'https://pokeapi.co/api/v2/'
response = requests.get(url + 'pokemon/tyranitar')
print(f"Request returned {response.status_code} : '{response.reason}'") #just to confirm 200 #here I want to know if I was successful
# in retrieving the data from the pokeAPI server
```

```python
response.json() #the data is in json format
```

```python
pokemon = response.json() #created a new variable to import the pokemon data
pokemon #printed the imported data
```

What is the Tyranitar's Types?

```python
pokemon['types'] #I want to extract the type attributes from the data
types = [] #new empty list to save output

for i in pokemon['types']: #I created a for loop that would look for types keys in all the data
    types.append(i['type']['name']) #I appended to the empty types list the name of each individual type 
types #output of Tyranitar names and types

print ('Tyranitar\'s types are', types[0], 'and', types[1]) #created a clean out and indexed the placement of my two types
```

What is the Tyranitar's Abilities?

```python
pokemon['abilities']
abilities = []

for k in pokemon['abilities']:
    abilities.append(k['ability']['name'])
abilities

print('The Tyranitar\'s abilities are', abilities[0], 'and', abilities[1])
```

### Question 3

Use the [Open Air Quality API](https://docs.openaq.org/)

Hint: typically queries need a ? in the url. Note that cities that need a space use a %20 between them (ie Los%20Angeles). Make sure that when you request information about a measurement that you are using a url that can do this for you.

First find out what the parameter id is of 'Particulate matter less than 25 micrometers in diameter.'

Then, find out how much 'Particulate matter less than 25 micrometers in diameter' there is in:
- Beijing
- Addis Ababa
- Dushanbe

Bonus points if you can aggregate this information in one table.

You must use python and apply the GET method via their API and display this information.



```python
PMurl = url+'v1/parameters?parameters=pm25' #given the new parameters, I added pm = 25 to the url

response = requests.get(PMurl)

pm25=response.json()['results'][0] #I retrieved the results from an array dictionary
#I indexed it at 0 because there was only one result
print(pm25)
```

```python
cities = ['Beijing', 'Addis%20Ababa', 'Dushanbe']
parameter = 'pm25'

for city in cities:
    cityurl = f'https://api.openaq.org/v1/measurements?city={city}&parameter={parameter}&limit=1'
    response = requests.get(cityurl)
    data = response.json()['results']
    
    if data:  # Check if the 'results' list is not empty
        print(f"{city}: {data[0]['value']}")
    else:
        print(f"No data found for {city}")
```
